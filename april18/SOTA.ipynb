```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["Google_API"]=os.getenv("Google_GenAI_API")
os.environ["OpenAI_API"]=os.getenv("OPENAI_API")
os.environ["HF_API"]=os.getenv("HUGGINGFACE_API")
```

```python
from sentence_transformers import SentenceTransformer
```

```python
open_source_embedding_model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sample_text="Syed Noor Mujassum"
```

```python
embedding=open_source_embedding_model.encode(sample_text)
```

```python
len(embedding)
```

```python
embedding.shape
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_openai import OpenAIEmbeddings
```

```python
google_embedding_model=GoogleGenerativeAIEmbeddings(model="gemini-embedding-001",api_key=os.environ["Google_API"])
```

```python
google_embedding_model.embed_query("How are you?")
```

```python
openai_embedding=OpenAIEmbeddings(api_key=os.environ["OpenAI_API"],model="text-embedding-3-small")
```

```python
openai_Embeddings_total=openai_embedding.embed_query(text="How are you ?")
```
