## Average Word2Vec

```python
sentence="Hi I am Syed Noor I love exploring AI"
```

```python
tokens=sentence.lower().split()
```

```python
tokens
```

```python
word_vectors=[]
```

```python
for word in tokens:
    print(word)
    word_vectors.append(model[word])
```

```python
word_vectors
```

```python
import numpy as np

vectors_sentence=np.mean(word_vectors,axis=0)
```
