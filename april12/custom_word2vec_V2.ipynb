```python
import os
import pandas as pd
import numpy as np
import nltk
import gensim
from nltk import sent_tokenize,word_tokenize
from nltk import PorterStemmer,WordNetLemmatizer,SnowballStemmer
from gensim.utils import simple_preprocess
```

```python
sentences=[]
with open ("./data/ai.txt",encoding="unicode_escape") as f:
    corpus=f.read()
    raw_sent=sent_tokenize(corpus)
    for sent in raw_sent:
        sentences.append(simple_preprocess(sent))
```

```python
custom_model=gensim.models.Word2Vec(window=20,min_count=2,vector_size=300,sg=1)
```

```python
custom_model.build_vocab(sentences)
```

```python
custom_model.train(sentences,total_examples=custom_model.corpus_count,epochs=5) #epochs = hyperparameter value
```
