```python
import os
import pandas as pd
import numpy as np
import nltk
import gensim
nltk.download("all")
```

```python
from nltk import sent_tokenize,word_tokenize
from nltk import PorterStemmer,WordNetLemmatizer,SnowballStemmer
from gensim.utils import simple_preprocess
```

```python
for file in os.listdir("./data/"):
    print(file)
```

```python
sentences=[]
for file in os.listdir("./data/"):
    file_path=os.path.join("./data/",file)
    with open (file_path,encoding="unicode_escape") as f:
        corpus=f.read()
        #sentences.append(corpus)
        raw_sent=sent_tokenize(corpus)
        for sent in raw_sent:
            sentences.append(simple_preprocess(sent))
    
```

```python
custom_model=gensim.models.Word2Vec(window=10,min_count=2,vector_size=150)

#all these values are hyper parameters

# Word2Vec(sg=0) Performs CBOW by default
# Word2Vec(sg=1) Performs skipgram
```

```python
custom_model.build_vocab(sentences)
```

```python
total_examples=custom_model.corpus_count
print(total_examples) # total number of sentences in data
```

```python
custom_model.train(sentences,total_examples=custom_model.corpus_count,epochs=5) #epochs = hyperparameter value
```
