# Tokenization
- all the individual words present in a sentence or corpus

- process of converting paragraphs to sentence and sentences to tokens

```python
sentence="I am Syed Noor Mujassum, I am a Senior Data Engineer. I love exploring new places."
```

```python
sentences=sentence.lower().split('.')
```

```python
tokens=str(sentences)
```

```python
print(tokens.split())
```

```python
from nltk.tokenize import sent_tokenize # converts paragraph to sentences
```

```python
corpus="I am Syed Noor Mujassum, I am a Senior Data Engineer. I love exploring new places."
```

```python
# import nltk
# nltk.download('punkt_tab')
documents=sent_tokenize(corpus)
```

```python
for sentence in documents:
    print(sentence)
```

```python
from nltk.tokenize import word_tokenize # convert paragraph or sentence to words
word_tokenize(corpus)
```

```python
#wordpunct_tokenize can be used to tokenize the 's in the tokens

from nltk.tokenize import wordpunct_tokenize, TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
