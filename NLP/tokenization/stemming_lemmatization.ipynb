# Stemming
- stemming is a processing of reducing a word to its word stem. stemming is important in NLU and NLP

```python
words=["eating","eaten","writing","write","written","programming","programs","fairly","sportly","goes","going"]
```

# Porter Stemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
[print(word+"--->"+stemmer.stem(word))for word in words]
```

```python
stemmer.stem("congratulations")
```

```python
# smetimes stemming changes the context or meaning of words. one of the biggest draw backs of stemming. for which lemmatization is applied
```

# RegexpStemmerClass

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eatable')
```

```python
reg_stemmer.stem('congratulations')
```

```python
reg_stemmer.stem("writing")
```

# Snowball Stemmer
- stemming technique with better accuracy than porter stemmer 

```python
from nltk.stem import SnowballStemmer
```

```python
snow_stemmer=SnowballStemmer(language='english')
```

```python
print("Output of Snow Ball Stemmer\n")
[print(word+" ----> "+snow_stemmer.stem(word)) for word in words]
```

```python
print("Output of Porter Stemmer\n")
[print(word+"--->"+stemmer.stem(word))for word in words]
```

# Lemmatization
- Lemmatization overcomes problems of stemming 
- Wordnet lemmatizer technique is used
- Lemmatizing is similar to stemming but output is lemma which is a root word rather than word stem
- Lemmatization takes longer time than Stemming as it compares agains wordnet corpus and uses morphy() function

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
# words=["eating","eaten","writing","write","written","programming","programs","fairly","sportly"]
```

```python
print("Out put of Lemmatization \n")
[print(word+ " -------> " +lemmatizer.lemmatize(word,pos='v')) for word in words]
```
